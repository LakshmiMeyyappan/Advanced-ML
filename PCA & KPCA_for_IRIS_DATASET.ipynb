```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, KernelPCA

# Load data
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (linear)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply Kernel PCA (nonlinear)
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.2)
X_kpca = kpca.fit_transform(X_scaled)

# Plotting function
def plot_components(X_proj, title):
    plt.figure(figsize=(6, 5))
    for i, label in enumerate(np.unique(y)):
        plt.scatter(X_proj[y == label, 0], X_proj[y == label, 1], label=target_names[label])
    plt.title(title)
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Show both projections
plot_components(X_pca, "PCA: Linear Components")
plot_components(X_kpca, "Kernel PCA (RBF): Nonlinear Components")
```
